# Null Model Baseline for Hurricane Track Prediction

Simple persistence baseline: uses velocity from last two observations to predict forward.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
import warnings
warnings.filterwarnings('ignore')

SAVE_DIR = 'figures/'
DATA_DIR = 'data/'
os.makedirs(SAVE_DIR, exist_ok=True)
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
df = pd.read_pickle("data/hurricane_paths_processed.pkl")
df = df.sort_values(['sid', 'iso_time']).reset_index(drop=True)

if 'x_km' not in df.columns:
    raise ValueError("Metric coordinates not found. Run feature engineering first.")

print(f"Loaded {len(df):,} observations from {df['sid'].nunique():,} storms")
```

```python
with open('data/train_test_splits.pkl', 'rb') as f:
    split_data = pickle.load(f)

train_storms = split_data['train_storms']
test_storms = split_data['test_storms']
val_storms = split_data['val_storms']

valid_storms = df[df.groupby('sid')['sid'].transform('count') >= 3]['sid'].unique()
df_modeling = df[df['sid'].isin(valid_storms)].copy()

df_val = df_modeling[df_modeling['sid'].isin(val_storms)].copy()
df_test = df_modeling[df_modeling['sid'].isin(test_storms)].copy()
```

## Null Model Implementation


```python
def null_model_forecast(storm_data, t0_idx, lead_steps):
    """Simple persistence: use velocity from t0-1 and t0 to predict forward"""
    storm_data = storm_data.sort_values('iso_time').reset_index(drop=True)
    
    if t0_idx < 1 or t0_idx >= len(storm_data):
        return None
    
    vx = storm_data['x_km'].iloc[t0_idx] - storm_data['x_km'].iloc[t0_idx - 1]
    vy = storm_data['y_km'].iloc[t0_idx] - storm_data['y_km'].iloc[t0_idx - 1]
    
    x_current = storm_data['x_km'].iloc[t0_idx]
    y_current = storm_data['y_km'].iloc[t0_idx]
    
    forecasts = []
    true_observations = []
    
    for step in range(1, lead_steps + 1):
        t_target = t0_idx + step
        if t_target >= len(storm_data):
            break
        
        x_pred = x_current + vx
        y_pred = y_current + vy
        forecasts.append([x_pred, y_pred])
        
        true_obs = np.array([storm_data['x_km'].iloc[t_target], storm_data['y_km'].iloc[t_target]])
        true_observations.append(true_obs)
        
        x_current = x_pred
        y_current = y_pred
    
    forecasts = np.array(forecasts)
    true_observations = np.array(true_observations)
    errors = forecasts - true_observations
    distances = np.sqrt(errors[:, 0]**2 + errors[:, 1]**2)
    
    return {
        'forecasts': forecasts,
        'true_observations': true_observations,
        'distances_km': distances,
        'lead_times_hours': np.arange(6, (len(forecasts) + 1) * 6, 6)
    }
```

```python
def evaluate_null_model_sliding_origins(df_eval, storm_ids, lead_times_hours=[6, 12, 24, 48, 72],
                                        min_origin_idx=5, max_origins_per_storm=5):
    """Evaluate null model using sliding origins"""
    all_results = []
    
    for sid in storm_ids:
        storm = df_eval[df_eval['sid'] == sid].sort_values('iso_time').reset_index(drop=True)
        
        if len(storm) < min_origin_idx + max(lead_times_hours) // 6 + 1:
            continue
        
        max_lead_steps = max(lead_times_hours) // 6
        valid_origins = range(min_origin_idx, len(storm) - max_lead_steps)
        step = max(1, len(valid_origins) // max_origins_per_storm)
        valid_origins = list(valid_origins[::step])
        
        for t0_idx in valid_origins:
            forecast_result = null_model_forecast(storm, t0_idx, max_lead_steps)
            if forecast_result is None:
                continue
            
            for lead_hours in lead_times_hours:
                lead_steps = lead_hours // 6
                if lead_steps <= len(forecast_result['distances_km']):
                    all_results.append({
                        'storm_id': sid,
                        'origin_idx': t0_idx,
                        'lead_time_hours': lead_hours,
                        'error_km': forecast_result['distances_km'][lead_steps - 1]
                    })
    
    return pd.DataFrame(all_results)

def summarize_lead_time_errors(results_df):
    """Summarize errors by lead time"""
    return results_df.groupby('lead_time_hours')['error_km'].agg([
        ('mean', 'mean'),
        ('median', 'median'),
        ('rmse', lambda x: np.sqrt(np.mean(x**2))),
        ('count', 'count')
    ]).round(2)
```

```python
min_storm_length = 5 + (72 // 6) + 1
val_storm_ids = df_val[df_val.groupby('sid')['sid'].transform('count') >= min_storm_length]['sid'].unique()
test_storm_ids = df_test[df_test.groupby('sid')['sid'].transform('count') >= min_storm_length]['sid'].unique()

val_null_results = evaluate_null_model_sliding_origins(df_val, val_storm_ids)
test_null_results = evaluate_null_model_sliding_origins(df_test, test_storm_ids)

print("Validation Set Results:")
print(summarize_lead_time_errors(val_null_results))
print("\nTest Set Results:")
print(summarize_lead_time_errors(test_null_results))
```

## Comparison with Kalman Filter


```python
kf_file = 'data/sliding_results_final.pkl'
if not os.path.exists(kf_file):
    kf_file = 'data/kalman_results_sliding_results_final.pkl'

if os.path.exists(kf_file):
    kf_results = pd.read_pickle(kf_file)
    if 'storm_id' in kf_results.columns:
        kf_results = kf_results.rename(columns={'storm_id': 'sid'})
    
    null_summary = summarize_lead_time_errors(test_null_results)
    kf_summary = summarize_lead_time_errors(kf_results)
    
    comparison = pd.DataFrame({
        'Null Model RMSE': null_summary['rmse'],
        'KF RMSE': kf_summary['rmse'],
        'Null Model Mean': null_summary['mean'],
        'KF Mean': kf_summary['mean'],
        'Improvement %': ((null_summary['rmse'] - kf_summary['rmse']) / null_summary['rmse'] * 100)
    })
    
    print("Comparison (Test Set):")
    print(comparison)
    
    comparison.to_pickle('data/null_vs_kf_comparison.pkl')
else:
    print("KF results not found. Run kalman_filter.ipynb first.")
```

## Visualizations


```python
if 'kf_results' in globals() and len(test_null_results) > 0:
    fig, ax = plt.subplots(figsize=(10, 6))
    
    lead_times = sorted(test_null_results['lead_time_hours'].unique())
    null_summary = summarize_lead_time_errors(test_null_results)
    kf_summary = summarize_lead_time_errors(kf_results)
    
    null_rmse = [null_summary.loc[lt, 'rmse'] for lt in lead_times]
    kf_rmse = [kf_summary.loc[lt, 'rmse'] for lt in lead_times]
    
    ax.plot(lead_times, null_rmse, 'o-', linewidth=2, markersize=8, label='Null Model', color='red')
    ax.plot(lead_times, kf_rmse, 's-', linewidth=2, markersize=8, label='Kalman Filter', color='blue')
    
    ax.set_xlabel('Lead Time (hours)', fontsize=12)
    ax.set_ylabel('RMSE (km)', fontsize=12)
    ax.set_title('Forecast Accuracy Comparison: Null Model vs Kalman Filter', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=11)
    
    for i, lt in enumerate(lead_times):
        improvement = ((null_rmse[i] - kf_rmse[i]) / null_rmse[i] * 100)
        if improvement > 0:
            ax.annotate(f'{improvement:.1f}%', xy=(lt, kf_rmse[i]), xytext=(lt, kf_rmse[i] - 20),
                       fontsize=9, color='green', fontweight='bold', ha='center',
                       arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
    
    plt.tight_layout()
    plt.savefig(f'{SAVE_DIR}fig_null_vs_kf_rmse_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
if 'kf_results' in globals() and len(test_null_results) > 0:
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    for idx, lead_time in enumerate([6, 12, 24, 48, 72]):
        ax = axes[idx]
        null_errors = test_null_results[test_null_results['lead_time_hours'] == lead_time]['error_km'].values
        kf_errors = kf_results[kf_results['lead_time_hours'] == lead_time]['error_km'].values
        
        bins = np.linspace(0, min(max(null_errors.max(), kf_errors.max()), 500), 50)
        ax.hist(null_errors, bins=bins, alpha=0.6, label=f'Null Model (n={len(null_errors)})', 
               color='red', density=True)
        ax.hist(kf_errors, bins=bins, alpha=0.6, label=f'Kalman Filter (n={len(kf_errors)})', 
               color='blue', density=True)
        
        ax.set_xlabel('Forecast Error (km)', fontsize=10)
        ax.set_ylabel('Density', fontsize=10)
        ax.set_title(f'{lead_time}h Lead Time', fontsize=11, fontweight='bold')
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3)
        
        null_rmse = np.sqrt(np.mean(null_errors**2))
        kf_rmse = np.sqrt(np.mean(kf_errors**2))
        improvement = ((null_rmse - kf_rmse) / null_rmse * 100)
        textstr = f'Null RMSE: {null_rmse:.1f} km\nKF RMSE: {kf_rmse:.1f} km\nImprovement: {improvement:+.1f}%'
        ax.text(0.65, 0.95, textstr, transform=ax.transAxes, fontsize=8,
               verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    axes[5].axis('off')
    plt.suptitle('Error Distribution Comparison: Null Model vs Kalman Filter', 
                fontsize=14, fontweight='bold', y=0.995)
    plt.tight_layout()
    plt.savefig(f'{SAVE_DIR}fig_null_vs_kf_error_distributions.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
if len(test_null_results) > 0:
    example_storm_id = test_null_results['storm_id'].iloc[0]
    example_storm = df_test[df_test['sid'] == example_storm_id].sort_values('iso_time').reset_index(drop=True)
    t0_idx = len(example_storm) // 2
    null_result = null_model_forecast(example_storm, t0_idx, 8)
    
    fig, ax = plt.subplots(figsize=(12, 8))
    
    if 'lat_ref' in example_storm.columns:
        lat_ref = example_storm['lat_ref'].iloc[0]
        lon_ref = example_storm['lon_ref'].iloc[0]
        
        def metric_to_latlon(x_km, y_km):
            R_EARTH = 111.0
            lat = lat_ref + y_km / R_EARTH
            lon = lon_ref + x_km / (R_EARTH * np.cos(np.radians(lat_ref)))
            return lat, lon
        
        actual_lat, actual_lon = metric_to_latlon(example_storm['x_km'].values, example_storm['y_km'].values)
        forecast_lat, forecast_lon = metric_to_latlon(null_result['forecasts'][:, 0], null_result['forecasts'][:, 1])
        true_lat, true_lon = metric_to_latlon(null_result['true_observations'][:, 0], null_result['true_observations'][:, 1])
        
        ax.plot(actual_lon, actual_lat, 'b-', linewidth=2.5, label='Actual track', alpha=0.7)
        ax.plot(forecast_lon, forecast_lat, 'r--', linewidth=2, label='Null model forecast', alpha=0.8)
        ax.scatter(true_lon, true_lat, c='green', s=50, marker='o', label='True positions', alpha=0.6)
        ax.scatter(actual_lon[t0_idx], actual_lat[t0_idx], c='black', s=150, marker='*', 
                   label='Forecast origin', zorder=5, edgecolors='white', linewidths=1)
        ax.set_xlabel('Longitude', fontsize=12)
        ax.set_ylabel('Latitude', fontsize=12)
    else:
        ax.plot(example_storm['x_km'].values, example_storm['y_km'].values, 
               'b-', linewidth=2.5, label='Actual track', alpha=0.7)
        ax.plot(null_result['forecasts'][:, 0], null_result['forecasts'][:, 1], 
               'r--', linewidth=2, label='Null model forecast', alpha=0.8)
        ax.scatter(null_result['true_observations'][:, 0], null_result['true_observations'][:, 1],
                  c='green', s=50, marker='o', label='True positions', alpha=0.6)
        ax.scatter(example_storm['x_km'].iloc[t0_idx], example_storm['y_km'].iloc[t0_idx],
                  c='black', s=150, marker='*', label='Forecast origin', zorder=5)
        ax.set_xlabel('x (km)', fontsize=12)
        ax.set_ylabel('y (km)', fontsize=12)
    
    mean_error = np.mean(null_result['distances_km'])
    rmse = np.sqrt(np.mean(null_result['distances_km']**2))
    textstr = f'Mean Error: {mean_error:.1f} km\nRMSE: {rmse:.1f} km'
    ax.text(0.02, 0.98, textstr, transform=ax.transAxes, fontsize=10,
           verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
    
    ax.set_title(f'Example Storm Track: Null Model Forecast\nStorm {example_storm_id}', 
                fontsize=13, fontweight='bold')
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig(f'{SAVE_DIR}fig_null_model_example_track.png', dpi=300, bbox_inches='tight')
    plt.show()
```
